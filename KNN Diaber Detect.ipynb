```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
diabet_df = pd.read_csv("diabetes.csv")
diabet_df.head(14)
```

```python
saglykly_hassalar = diabet_df[diabet_df.Outcome == 0]
saglyklyksyz_hassalar = diabet_df[diabet_df.Outcome == 1]
```

```python
saglyk_sany =  len(saglykly_hassalar)
saglyksyz_sany = len(saglyklyksyz_hassalar) 
```

```python
plt.xlabel("Ady")
plt.ylabel("Sany")
plt.bar(["Diabets","Not Diabets"],[saglyksyz_sany,saglyk_sany],color="black")
plt.show()
```

```python
plt.scatter(saglykly_hassalar.Age,saglykly_hassalar.Insulin, color="green", label="Saglykly Hassalar",alpha=0.4)
plt.scatter(saglyklyksyz_hassalar.Age,saglyklyksyz_hassalar.Insulin, color="red",label="Saglyklyksyz Hassalar",alpha=0.4)
plt.legend()
plt.show()
```

```python
plt.scatter(saglykly_hassalar.Age,saglykly_hassalar.Glucose, color="green", label="Saglykly Hassalar",alpha=0.4)
plt.scatter(saglyklyksyz_hassalar.Age,saglyklyksyz_hassalar.Glucose, color="red",label="Saglyklyksyz Hassalar",alpha=0.4)
plt.legend()
plt.show()
```

```python
y = diabet_df.Outcome.values
```

```python
x = diabet_df.drop(["Outcome"],axis=1)
```

```python
x_normal = (x - np.min(x)) / (np.max(x)-np.min(x))
```

```python
print("NOrmalization\n",x_normal.head())
print("NOt  \n",x.head())
```

```python
x_train ,x_test,y_train,y_test = train_test_split(x_normal,y,test_size = 0.2,random_state=1)
```

```python
ModelKNN = KNeighborsClassifier(n_neighbors=7)
ModelKNN.fit(x_train,y_train)
prediction = ModelKNN.predict(x_test)
print("Knn 3 n bilrn dogrulyk scory ",ModelKNN.score(x_test,y_test))
```

```python
sanayjy = 1 
for k in range(1,11):
    ModelKNNN = KNeighborsClassifier(n_neighbors=k)
    ModelKNNN.fit(x_train,y_train)
    prediction = ModelKNNN.predict(x_test)
    kk = ModelKNNN.score(x_test,y_test)
    print(sanayjy," Knn ",sanayjy," n bilen dogrulyk scory ",kk)
    sanayjy += 1
    
```
